```python
import csv
from prettytable import PrettyTable as PT
```

```python
def get_data(Filename):
    '''this function takes in the filename of csv and returns the lists of lists'''
    with open(Filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        Station = []
        Date = []
        maxTemp = []
        minTemp = []
        awnd = []
        Precip = []
        Data = []
    
        for row in csv_reader:
            if line_count == 0:
                #print(f'Column names are {", ".join(row)}')
                line_count += 1
            else:
                #print(f'\t {row[0]} | {row[1]} | {row[2]} {row[3]} | {row[4]} | {row[5]}')
                Station.append(row[0])
                Date.append(row[1])
                maxTemp.append(row[2])
                minTemp.append(row[3])
                awnd.append(row[4])
                Precip.append(row[5])
                line_count += 1
            
        Data.append(Station)
        Data.append(Date)
        Data.append(maxTemp)
        Data.append(minTemp)
        Data.append(awnd)
        Data.append(Precip)
    
    return Data
```

```python
def print_data(list):
    '''this function takes in the lists of lists and returns a pretty table of all the data.'''
    temp_list = []
    Total_Days = 0
    
    for item in list:
        Total_Days += len(item)
    Total_Days = Total_Days/len(list)
    
    myTable = PT(["STATION", "DATE", "TMAX", "TMIN", "AWND", "PRCP"])
    
    count = 0
    while(Total_Days > count):
        temp_list.clear()
        
        for item in list:
            temp_list.append(item[count]) 
        myTable.add_row(temp_list)
        count = count + 1
        
    myTable.add_row(calc_averages(list))
    
    return myTable
```

```python
def calc_averages(list):
    column_avgs = []
    Maxtemp = []
    Mintemp = []
    PRCP = []
    temp_int = []
    final_data = ["",""]
    count = 0
    avg = 0
    
    Maxtemp = list[2]
    Mintemp = list[3]
    PRCP = list[5]
    
    for item in Maxtemp:
        count = count + float(item)
        
    avg = count/len(Maxtemp)
    final_data.append(round(avg,2))
    
    count = 0
    for item in Mintemp:
        count = count + float(item)
    
    avg = count/len(Mintemp)
    final_data.append(round(avg,2))
    
    final_data.append("")
    
    count = 0
    for item in PRCP:
        count = count + float(item)
        
    avg = count/len(PRCP)
    final_data.append(round(avg,2))
    return final_data
```

```python
def main():
    '''This executes the program loop and runs all other functions'''
    
    print("Welcome to the Weather Reporter!")
    file = input("Enter name of the weather data file: ")
    
    user_input = "0"
    
    while(user_input != "-1"):
        print("Weather Reporter \n =============================")
        print("1. Print Table (Entire Year)")
        print("2. Print Table (by Quarter)")
        print("3. Print Table (by Month)")
        print("\n")
        
        user_input = input("Choose option (-1 to quit): ")
        data = get_data(file)
        
        if(user_input == "1"):
            print(print_data(data))
            
        if(user_input == "2"):
            print(get_quarter_data(data))
        
        if(user_input == "3"):
            print(get_month_data(data))
    print("Thanks for using Weather Reporter")
```

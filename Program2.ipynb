```python
from prettytable import PrettyTable as PT
import sys
```

```python
def derive_filename(Poemname):
    '''This function takes in Poemname and derives the filename from the name of poem. returns filename'''
    Filename = ""
    tempName = []
    count = 0
    
    while(count < len(Poemname)):
        tempName.append(Poemname[count])
        #print(tempName[count])
        count = count + 1
    
    count = 0

    while(count < len(tempName)):
        if(tempName[count] == " "):
            tempName[count] = "_"
        if(tempName[count].isupper()):
            tempName[count] = tempName[count].lower()
        Filename = Filename + tempName[count]
        count = count + 1
    
    Filename = Filename + ".txt"
    return Filename
```

```python
def count_stanzas(file):
    '''This function takes in the filename and returns the number of stanzas'''
    stanzas = 0
    total_indents = 0 
    
    r = open(file, "r")
    Poem = r.readlines()
    
    count = 0
    while(len(Poem) > count):        
        if(Poem[count] == '\n'):
            stanzas = stanzas + 1
        count = count + 1
        
    return stanzas
```

```python
def count_lines(file):
    '''this function takes in the filename and returns the number of lines'''
    lines = 0
    
    r = open(file, "r")
    Poem = r.readlines()
    
    count = 0
    while(len(Poem) > count):
        if(Poem[count] == '\n'):
            pass
        else:
            lines = lines + 1
        count = count + 1
        
    return lines
```

```python
def average_words_line(file):
    '''This function takes in the filename and returns the average number of words per line'''
    line_total = count_lines(file)
    word_total = 0
    
    r = open(file, "r")
    Poem = r.readlines()
    
    count = 0
    while(len(Poem)> count):
        for word in Poem[count].split():
            word_total = word_total + 1
        count = count + 1
        
    avg_words_line = round(word_total/line_total, 2)
    return avg_words_line
```

```python
def average_words_stanza(file):
    '''this function reads the file and calculates word total and uses count_stanzas to return the average words per stanza'''
    stanza_total = count_stanzas(file)
    word_total = 0

    r = open(file, "r")
    Poem = r.readlines()
    
    count = 0
    while(len(Poem)> count):
        for word in Poem[count].split():
            word_total = word_total + 1
        count = count + 1
        
    avg_words_stanza = round(word_total/stanza_total, 2)
    return avg_words_stanza
```

```python
def average_word_length(file):
    '''this function reads the file and calculates word and character total to return the average characters per word'''
    word_total = 0
    character_total = 0
    
    r = open(file, "r")
    Poem = r.readlines()    
    
    count = 0
    while(len(Poem)> count):
        for word in Poem[count].split():
            word_total = word_total + 1
            for character in word:
                character_total = character_total + 1
        count = count + 1
        
    average_word_length = round((character_total/word_total),2)
    
    return average_word_length 
```

```python
def print_poem(file, Title):
    '''this function adds the title to the beginning of the poem and prints it out'''
    r = open(file, "r")
    Poem = r.readlines()
    
    total_Poem = [Title.upper()]
    total_Poem = total_Poem + Poem
    
    for item in total_Poem:
        print(item)
```

```python
def process_poem(Title, number_of_stanzas, number_of_lines, avg_word_line, avg_word_stanza, avg_word_length):
    '''this function creates a dictionary of all functions related to the Poem: Title, number_of_stanzas, number_of_lines, avg_word_line, avg_word_stanza, avg_word_length'''
    
    thisdict = {
        "Poem": Title,
        "Number of Stanzas": number_of_stanzas,
        "Number of Lines": number_of_lines,
        "Avg Words/Line": avg_word_line,
        "Avg Words/Stanza": avg_word_stanza,
        "Avg Word Length": avg_word_length
    }
    
    return thisdict
```

```python
def print_poem_stats(dict):
    '''this function takes in the dictionary and creates/prints a pretty table of the data within the dictionary'''
    
    myTable = PT(["Poem", "Number of Stanzas", "Number of Lines", "Avg Words/Line", "Avg Words/Stanza", "Avg Word Length"])
    row = [dict["Poem"], dict["Number of Stanzas"], dict["Number of Lines"], dict["Avg Words/Line"], dict["Avg Words/Stanza"], dict["Avg Word Length"]]
    myTable.add_row(row)
    
    print(myTable)
    
```

```python
def main():
    
    '''this is the main loop involves all other functions and user input'''

    user_input = input("Enter poem name (-1 to quit):")

    while user_input != -1:
        Title = user_input
    
        file = derive_filename(Title)
        stanzas = count_stanzas(file)
        lines = count_lines(file)
        avg_words_line = average_words_line(file)
        avg_words_stanza = average_words_stanza(file)
        avg_word_length = average_word_length(file)
        dict = process_poem(Title.upper(), stanzas, lines, avg_words_line, avg_words_stanza, avg_word_length)
        print_poem_stats(dict)
    
        print_poem(file,Title)
        print("\n")
    
        user_input = input("Enter poem name (-1 to quit):")
        if(user_input == "-1"):
            user_input = int(user_input)
```

```python
main()
```

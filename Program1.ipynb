```python
import math
from prettytable import PrettyTable as PT
```

```python
def cylinderVolume(r,h):
    '''Takes 2 Parameters (radius and height) and returns Volume'''
    Volume = math.pi*math.pow(r,2)*h
    return Volume
```

```python
def cylinderSurface(r,h):
    '''Takes 2 Parameters (radius and height) and returns Surface Area'''
    SA = (2*math.pi*r*h)+(2*math.pi*math.pow(r,2))
    return SA
```

```python
def coneVolume(r,h):
    '''Takes 2 Parameters (radius and height) and returns the Volume'''
    Volume = (math.pi*math.pow(r,2)*(h/3))
    return Volume
```

```python
def coneSurface(r,h):
    '''Takes 2 Parameters (radius and height) and Returns Surface Area'''
    SA = (math.pi*r*(r+math.sqrt(math.pow(h,2)+math.pow(r,2))))
    return SA
```

```python
def display_menu():
    '''displays a menu for user to know their function options'''
    r = "     Geometry Calculator\n============================= "
    r = r + ("\n1. Calculate the volume of a cylinder\n2. Calculate the suface area of a cylinder\n3.Calculate the volume of a cone")
    r = r+ ("\n4. Calculate the surface area of a cone\n5. Print Table\n6. Save Data")
    return r
```

```python
def print_table(data):
    '''creates a neat table of data and displays it for the user'''
    print(data)
    myTable = PT(["Measurement", "Radius", "Height", "Value"])
    count = len(data)
    i = 0
    while(i < count):
        myTable.add_row(data[i])
        i = i + 1
    print(myTable)
```

```python
def save_data(data, filename):
    '''Creates new file of input name and writes the list of calculations into that file'''
    print(filename)
    f = open(filename, "x")
    for item in data:
        for i in item:
            f.write(str(i) + "\n")
        
```
